```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = 'A3_seg.npz'
```

```python
with np.load(data_path) as data:
    time = data['time']
    im_cardio = data['cardio']
    im_watershed = data['cardio_watershed']
    im_cover = data['cardio_cover']
    im_pred = data['cardio_pred']
    mic = data['mic']
    mic_singular = data['mic_singular']
    marker = data['marker']
    marker_singular = data['marker_singular']
    
    well = data['well']             # biosensor
    im_markers = data['im_markers'] # marker
    im_mic = data['im_mic']         # microscope
```

```python
print('tags: ', data.files)
print('time dim: \t\t\t\t\t', time.shape, )
print('bioszenzor 8x8px képek 772 idődim \t\t', im_cardio.shape, im_watershed.shape, im_cover.shape, im_pred.shape)
print('8x8 db 412x412 px img??: \t\t\t', mic.shape, marker.shape, mic_singular.shape, marker_singular.shape,)
print('biosensor img: \t\t\t\t\t', well.shape)
print('microscope img (3 channel) and the markings: \t', im_markers.shape, im_mic.shape)
```

```python
fig, axs = plt.subplots(1, 4, figsize=(10, 10))

i = 25
t = 350

axs[0].imshow(im_cardio[i][t])
axs[1].imshow(im_watershed[i][t])
axs[2].imshow(im_cover[i][t])
axs[3].imshow(im_pred[i][t])

plt.show()
```
