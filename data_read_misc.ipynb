```python
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import pickle 
```

```python
fibronectin = 'fibronectin/*/*.npz'
noncoated = 'noncoated/*/*.npz'
```

```python
fibronectin_npz_files = glob(fibronectin)
noncoated_npz_files = glob(noncoated)
```

```python
print(fibronectin_npz_files)
print(noncoated_npz_files)
```

```python
print(len(noncoated_npz_files), len(fibronectin_npz_files))
```

```python
file_number = 0
files = fibronectin_npz_files + noncoated_npz_files

for file in files:
    data = np.load(file)

    new_data = {'old_filename': file, 'biosensor': data['well'], 'mask': data['im_markers']}
    filename = f'neccesary/{file_number}.npz'
    np.savez(filename, **new_data)
    file_number += 1
    
```

```python
all_files = glob('neccesary/*.npz')
print(all_files.__len__())
```

```python
# The code below saves only the neccecary data to a pickle but for the sake of batch dataloading the above code is used to make the minimalised dataset.

biosensor = []
microscope = []
masks = []
```

```python
for file in fibronectin_npz_files:
    with np.load(file) as data:
        biosensor.append(data['well'])
        microscope.append(data['im_mic'])
        masks.append(data['im_markers'])
```

```python
for file in noncoated_npz_files:
    with np.load(file) as data:
        biosensor.append(data['well'])
        microscope.append(data['im_mic'])
        masks.append(data['im_markers'])
```

```python
data_len = len(biosensor)
print(data_len)
```

```python
# Vannak maszkok amik nem négyzetek
for mask in masks:
    if mask.shape[0] != mask.shape[1]:
        print(mask.shape)
```

```python
all_npz = glob('neccesary/*.npz')
for file in all_npz:
    with np.load(file) as data:
        print(data['old_filename'] , data['biosensor'].shape, data['mask'].shape)
```

```python
# The biosensor length is not the same for all the files
avg_time = np.mean([bio.shape[0] for bio in biosensor])
std_time = np.std([bio.shape[0] for bio in biosensor])
min_time = np.min([bio.shape[0] for bio in biosensor])
max_time = np.max([bio.shape[0] for bio in biosensor])
print(avg_time, std_time, min_time, max_time)
```

```python
# The mask or the microscope image is not the same for all the files
avg_mask = np.mean([mask.shape[0] for mask in masks])
avg_mask = np.mean([mask.shape[0] for mask in masks])
std_mask = np.std([mask.shape[0] for mask in masks])
min_mask = np.min([mask.shape[0] for mask in masks])
max_mask = np.max([mask.shape[0] for mask in masks])
print(avg_mask, std_mask, min_mask, max_mask)
```

```python
with open('pickles/biosensor.pkl', 'wb') as out_file:
    pickle.dump(biosensor, out_file)
    
# with open('pickles/microscope.pkl', 'wb') as out_file:
#     pickle.dump(microscope, out_file)
    
with open('pickles/masks.pkl', 'wb') as out_file:
    pickle.dump(masks, out_file)
```
